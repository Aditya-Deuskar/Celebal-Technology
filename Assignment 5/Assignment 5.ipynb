```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
# we have 2 datasets in the provided data, so we will combine them into a single dataset and then perform the operations which are asked in the queestion.
'''so the test dataset does not have a column named "Survived" and this is present in the test dataset, so this may cause problems in the combined dataset
so we are gonna add the same column with 'NaN' values in the test data so that things become easy for us in the future'''

test["Survived"] = pd.NA
```

```python
#now we will combine the two datasets and store it into a new dataset called combined

combined = pd.concat([train, test], ignore_index=True)
combined.to_csv('combined.csv', index=False)
```

```python
#here we will handle the missing values in those columns

combined['Age'].fillna(combined['Age'].median(), inplace = True)
combined['Fare'].fillna(combined['Fare'].median(), inplace = True)

#as embarked is a categorical variable we will prefer using mode instead of median herea
combined['Embarked'].fillna(combined['Embarked'].mode()[0], inplace=True)

# Create a new feature 'HasCabin' indicating whether a cabin is known
combined['HasCabin'] = combined['Cabin'].apply(lambda x: 0 if pd.isna(x) else 1)

combined.drop('Cabin', axis=1, inplace=True)
```

```python
# here we will normalize the numerical features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
combined[['Age', 'Fare', 'SibSp', 'Parch']] = scaler.fit_transform(combined[['Age', 'Fare', 'SibSp', 'Parch']])
```

```python
#here we will encode categorical variables and normalize the data
combined = pd.get_dummies(combined, columns=['Sex', 'Embarked'], drop_first=True)
```

```python
#here we will perform feature engineering and add two more features 
combined['FamilySize'] = combined['SibSp'] + combined['Parch'] + 1
combined['IsAlone'] = combined['FamilySize'].apply(lambda x: 1 if x == 1 else 0)
```

```python
combined.columns
```

```python
# 1. Survival Rate by Passenger Class
plt.subplot(2, 2, 1)
sns.barplot(x='Pclass', y='Survived', data=combined)
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')

# 2. Gender Distribution
plt.subplot(2, 2, 2)
gender_counts = combined['Sex_male'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index,  autopct='%1.1f%%',startangle=140, colors=['lightblue', 'lightpink'])
plt.title('Gender Distribution (True if male)')

# 3. Age Distribution
plt.subplot(2, 2, 3)
sns.histplot(combined['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')

# 4. Survival Rate by Gender
plt.subplot(2, 2, 4)
sns.barplot(x='Sex_male', y='Survived', data=combined)
plt.title('Survival Rate by Gender')
plt.ylabel('Survival Rate')

plt.tight_layout()
plt.show()
```

```python
# now we will Separate the combined dataset back into training and test datasets as it was before we took it for preprocessing
train = combined[combined['Survived'].notna()]
test = combined[combined['Survived'].isna()]
```

```python
# Save the preprocessed datasets
train.to_csv('train_preprocessed.csv', index=False)
test.to_csv('test_preprocessed.csv', index=False)
```
