```python
from azure.ai.textanalytics import TextAnalyticsClient
from azure.core.credentials import AzureKeyCredential

# Replace these with your actual values
endpoint = "https://<your-resource-name>.cognitiveservices.azure.com/"
api_key = "<your-api-key>"

# Instantiate the client
client = TextAnalyticsClient(endpoint=endpoint, credential=AzureKeyCredential(api_key))


def authenticate_client(key, endpoint):
    """Authenticate the client with API key and endpoint."""
    ta_credential = AzureKeyCredential(key)
    text_analytics_client = TextAnalyticsClient(
        endpoint=endpoint, 
        credential=ta_credential
    )
    return text_analytics_client

def sentiment_analysis_example(client, reviews):
    """Perform sentiment analysis on movie reviews."""
    response = client.analyze_sentiment(documents=reviews)
    sentiments = []
    for idx, document in enumerate(response):
        print(f"Review {idx+1}:")
        print(f"\tSentiment: {document.sentiment}")
        print(f"\tConfidence Scores: Positive={document.confidence_scores.positive}; Neutral={document.confidence_scores.neutral}; Negative={document.confidence_scores.negative}")
        sentiments.append(document.sentiment)
    return sentiments

if __name__ == "__main__":
    # Replace with your Azure Text Analytics credentials
    api_key = "YOUR_AZURE_TEXT_ANALYTICS_API_KEY"
    endpoint = "YOUR_AZURE_TEXT_ANALYTICS_ENDPOINT"

    # Authenticate the client
    client = authenticate_client(api_key, endpoint)

    # Sample movie reviews for sentiment analysis
    movie_reviews = [
        "I absolutely loved this movie. The plot was engaging and the characters were well-developed.",
        "The movie was okay, but I expected more action.",
        "I didn't enjoy this movie at all. It was too slow and boring."
    ]

    # Perform sentiment analysis
    sentiment_analysis_example(client, movie_reviews)
```
